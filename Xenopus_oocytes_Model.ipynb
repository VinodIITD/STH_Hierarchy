```python
# Groebner basis for positive feedback oocyte model

# Define ring and variables
R = PolynomialRing(QQ, 'A')
A = R.gen()

# Parameters
Atot = 1      
f = 0.13      
n = 5         
K = 1         
kinact = 0.01 
stimulus = 1 

# Construct steady-state polynomial (multiplied both sides by denominator for polynomial form)
poly = stimulus * (K**n + A**n) - f * A**n * (Atot - A) - kinact * A * (K**n + A**n)

# Print the polynomial
print('Polynomial for Groebner basis:')
print(poly)

# Create ideal from polynomial
I = R.ideal([poly])

# Compute Groebner basis
gb = I.groebner_basis()
print('Groebner basis:')
print(gb)

# # Finding real roots
# roots = poly.roots(ring=RR, multiplicities=False)
# print('Real positive roots for A:')
# for r in roots:
#     if r > 0 and r < Atot:
#         print(r)
        
# # Get all roots with multiplicities in algebraic closure
# all_roots = poly.roots(multiplicities=True)

# print('All roots found (possibly including complex):')
# for r, mult in all_roots:
#     print(f'Root: {r}, multiplicity: {mult}')

# print('Filtered real positive roots in [0, Atot]:')
# for r, mult in all_roots:
#     if r.imag() == 0:  # check root is real
#         real_val = r.real()
#         if 0 < real_val < Atot:
#             print(real_val)   
```
